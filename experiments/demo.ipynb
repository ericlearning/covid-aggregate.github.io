```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
data = pd.DataFrame(
    data = {
        'variable_x': np.arange(0, 100),
        'variable_1': np.arange(0, 100) + np.random.uniform(-10, 10, 100),
        'variable_2': np.arange(200, 0, -2) + np.random.uniform(-30, 30, 100),
        'variable_3': np.arange(0, 150, 1.5) + np.random.uniform(-50, 50, 100)
    }
)

fig = px.scatter(
    data,
    x='variable_x',
    y=['variable_1', 'variable_2', 'variable_3'],
    title='random relationships',
)
fig.update_xaxes(title_text='variable_x')
fig.update_yaxes(title_text='variable_y')

fig.show()
```
